```python
import pygad
import numpy as np

num_generations = 100
sol_per_pop = 20

# Initialize a PyGAD instance
ga_instance = pygad.GA(gene_space=[0, 1],
                       num_generations=num_generations,
                       fitness_func=create_GA_fitness(data[1], distMat, tsp_heuristic="NN"),
                       num_parents_mating=20,
                       num_genes=len(data[1]),
                       sol_per_pop=sol_per_pop,
                       crossover_type="single_point",
                       mutation_type="random"
                       )
ga_instance.run()
```

```python
solution, solution_fitness, solution_idx = ga_instance.best_solution()
print("Parameters of the best solution : {solution}".format(solution=solution))
print("Number of hubs in the best solution : {one_count}".format(one_count=np.sum(solution)))
print("Fitness value of the best solution = {solution_fitness}".format(
    solution_fitness=solution_fitness))
ansViz(data[1], solution, distMat, tsp_heuristic="NN")
print("Cost of the best solution = {solution_cost}".format(solution_cost=GA_cost(data[1], solution, distMat)))
```
